```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

nav = webdriver.Chrome()
nav.get('https://www.google.com/')

nav.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('cotação dólar', Keys.ENTER)
dolar = nav.find_element(By.XPATH,'//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')

nav.get('https://www.google.com/')
nav.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('cotação euro', Keys.ENTER)
euro = nav.find_element(By.XPATH,'//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')

nav.get('https://www.google.com/')
nav.get('https://www.melhorcambio.com/ouro-hoje#:~:text=O%20valor%20do%20grama%20do,em%20R%24%20280%2C60.')
ouro = nav.find_element(By.XPATH,'//*[@id="comercial"]').get_attribute('value')
ouro = ouro.replace(',', '.')
nav.quit()
```

```python
import pandas as pd
from datetime import datetime
table = pd.read_excel('Products.xlsx')

table.loc[table['Moeda'] == 'Dólar', 'Cotação'] = float(dolar)
table.loc[table['Moeda'] == 'Euro', 'Cotação'] = float(euro)
table.loc[table['Moeda'] == 'Ouro', 'Cotação'] = float(ouro)
table.loc[0, 'Last update'] = datetime.now()

table['Preço de Compra'] = table['Preço Original'] * table['Cotação']
table['Preço de Venda'] = table['Preço de Compra'] * table['Margem']

table.to_excel('new products.xlsx', index=False)
```
